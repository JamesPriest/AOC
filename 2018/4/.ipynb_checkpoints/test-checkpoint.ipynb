```python
import datetime
```

```python
data = open('input.txt').read().split("\n")
```

```python
data = data[:-1]
```

```python
data[0]
```

```python
dates = [i.split(']')[0].strip('[') for i in data]
```

```python
dates = [datetime.datetime.strptime(i, "%Y-%m-%d %H:%M") for i in dates]
```

```python
combined = list(zip(data, dates))
```

```python
combined_sorted = sorted(combined, key=lambda x:x[1])
```

```python
guards = []
for i in combined_sorted:
    if "Guard" in i[0]:
        guard_no = i[0].split(' ')[3]
        guards.append(guard_no)
```

```python
len(set(guards))
```

```python
guards_num = [int(i.strip("#")) for i in guards]
```
